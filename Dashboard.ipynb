```python
#importing the required libraries
import pywedge as pw
import pandas as pd
```

```python
#reading the data set (Human Resources DataSet)
df=pd.read_csv('/content/hr_data.csv')
df.head()
```

```python
mc=pw.Pywedge_Charts(df,c=None,y='salary')
#here salary is the target variable
```

```python
#using make_charts() inbuilt function to make an interactive interface in which user can select the type of chart / plot and the depend and independent variables 
charts=mc.make_charts()
```
